```python
df=pd.read_csv('D:\\APSHAH\\backend\\powerpredictionfinal.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Convert the "DateTime" column to datetime format
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Filter data up to 2024-01-01 07:00:00
filtered_data = df[df['DateTime'] <= '2024-01-02 00:00:00']

# Plotting the line chart
plt.plot(filtered_data['DateTime'], filtered_data['Power generated by system | (MW)'])

# Adding labels and title
plt.xlabel('Date and Time')
plt.ylabel('Air temperature (C)')
plt.title('Power generated by system | (MW) (Up to 2024-01-02 00:00:00)')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```
