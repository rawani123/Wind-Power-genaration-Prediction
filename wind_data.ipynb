```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from scipy.stats import uniform
import pandas as pd

# Load the data
features = new_gridData_2[['p1', 'p2', 'p3', 'c1', 'c2', 'c3', 'year', 'month', 'day', 'hour', 'minute', 'seconds', 'milliseconds', 'g1', 'g2', 'g3']]
target = new_gridData_2['stability']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Define the parameter grid or distributions
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Regularization strength
    'penalty': ['l1', 'l2'],  # Penalty type
    'solver': ['liblinear', 'saga'],  # Solver algorithm
    'max_iter': [100, 200, 300]  # Maximum number of iterations
}

param_dist = {
    'C': uniform(loc=0, scale=4),  # Uniform distribution for regularization strength
    'penalty': ['l1', 'l2'],  # Penalty type
    'solver': ['liblinear', 'saga'],  # Solver algorithm
    'max_iter': [100, 200, 300]  # Maximum number of iterations
}

# Create a logistic regression classifier
logreg = LogisticRegression()

# Instantiate GridSearchCV or RandomizedSearchCV
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy', verbose=1)
random_search = RandomizedSearchCV(logreg, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy', random_state=42, verbose=1)

# Fit the search to the data
grid_search.fit(X_train, y_train)
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params_grid = grid_search.best_params_
best_params_random = random_search.best_params_
print("Best hyperparameters from grid search:", best_params_grid)
print("Best hyperparameters from random search:", best_params_random)

# Get the best models
best_model_grid = grid_search.best_estimator_
best_model_random = random_search.best_estimator_

# Evaluate the best models on the test set
test_accuracy_grid = best_model_grid.score(X_test, y_test)
test_accuracy_random = best_model_random.score(X_test, y_test)
print("Test accuracy of best model from grid search:", test_accuracy_grid)
print("Test accuracy of best model from random search:", test_accuracy_random)
```
